```python
import numpy as np
import time
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Generate a random adjacency matrix
n_V = 100
n_E = 500
n = n_E
adj_mtrx = np.zeros((n_V, n_V))

# with assigned random positive integer weights
while n != 0:
    i, j = np.random.randint(0, 99), np.random.randint(0, 99)
    if i != j and adj_mtrx[i, j] == 0:
        weight = np.random.randint(0, 99)
        adj_mtrx[i, j], adj_mtrx[j, i] = weight, weight
        n -= 1

print(f'Shape: {adj_mtrx.shape}, weight sum: {np.sum(adj_mtrx)}')
```

```python
# Transfer the matrix into an adjacency list
adj_lst = {i: [] for i in range(n_V)}
for k, v in adj_lst.items():
    for i in range(n_V):
        adj_lst[k].append([i, adj_mtrx[k, i]]) if adj_mtrx[k, i] != 0 else next
```

```python
lst = []
for k, v in adj_lst.items():
    lst.extend([[k, i[0], i[1]] for i in v])
```

```python
def visualize(list):
        G = nx.Graph()
        plt.figure(figsize=(20,15))
        plt.title(f'My random graph with {n_V} vertices and {n_E} edges',
                  fontsize=25)
        visualize = [i[0:2] for i in list]
        G.add_edges_from(visualize)
        nx.draw_networkx(G)
        plt.show()
```

```python
import numpy as np
import time
import networkx as nx
import matplotlib.pyplot as plt

# Bellman-Ford algorithm
def bellman_ford(adj_matrix, start_vertex):
    num_vertices = adj_matrix.shape[0]
    distances = [float('inf')] * num_vertices
    distances[start_vertex] = 0

    start_time = time.time()  # Record start time

    for _ in range(num_vertices - 1):
        for u in range(num_vertices):
            for v in range(num_vertices):
                if adj_matrix[u][v] != 0 and distances[u] + adj_matrix[u][v] < distances[v]:
                    distances[v] = distances[u] + adj_matrix[u][v]

    for u in range(num_vertices):
        for v in range(num_vertices):
            if adj_matrix[u][v] != 0 and distances[u] + adj_matrix[u][v] < distances[v]:
                print("Graph contains negative weight cycle")
                return

    end_time = time.time()  # Record end time

    print("Vertex Distance from Source")
    for i in range(num_vertices):
        print("{0}\t\t{1}".format(i, distances[i]))

    print(f"\nTime taken: {end_time - start_time:.5f} seconds")

    return distances


# Choose a random starting vertex
start_vertex = np.random.randint(0, n_V)

# Apply the Bellman-Ford algorithm
print("\nApplying Bellman-Ford algorithm:")
bellman_ford(adj_mtrx, start_vertex)
```

```python
import numpy as np
import time
import networkx as nx
import matplotlib.pyplot as plt

num_repeats = 10

# Initialize a dictionary to store shortest distances for each vertex
shortest_distances = {i: [] for i in range(n_V)}

# Repeat the experiment for the same starting vertex
for i in range(num_repeats):
    distances = bellman_ford(adj_mtrx, start_vertex)
    
    # Store the distances for each vertex
    for vertex, distance in enumerate(distances):
        shortest_distances[vertex].append(distance)

# Calculate the average shortest distance for each vertex
average_shortest_distances = {vertex: sum(distances) / num_repeats for vertex, distances in shortest_distances.items()}

# Print the average shortest distances for each vertex
print("\nAverage Shortest Distances:")
for vertex, distance in average_shortest_distances.items():
    print(f"Vertex {vertex}: {distance:.5f}")
```
